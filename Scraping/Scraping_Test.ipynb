## 2.1 Requests
Use the requests class to perform a HTTP GET request for extracting information.

````python
import requests

URL = 'https://www.monster.com/jobs/search/?q=Software-Developer&where=hong-kong'
page = requests.get(URL)
````

````python
import pprint
pp = pprint.PrettyPrinter()
````

**Static websites - return HTML**

As printed below.

````python
pp.pprint(page.content)
````

## 2.2 BeautifulSoup parse
Use BeautifulSoup to parse the page's content.

````python
soup = BeautifulSoup(page.content, 'html.parser')
````

````python
print(type(soup))
````

````python
results = soup.find(id='ResultsContainer')
````

````python
print(results)
````

## 2.3 Prettify
Use ```prettify``` function to align the `<div>` tags in a better manner.

````python
print(results.prettify())
````

## 2.4 Separate entries
We find that the `<section>` tag separates each job entry, so let's separate by that using `find_all` function.

````python
job_elems = results.find_all('section', class_='card-content')
````

````python
for job_elem in job_elems:
    print(job_elem, end='\n'*2)
````

Further separate by `<h2>`, `<div class="company">` and `<div class="location">` tags, which separates into job title, company name and location of job.

````python
for job_elem in job_elems:
    # Each job_elem is a new BeautifulSoup object.
    # You can use the same methods on it as you did before.
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    print(title_elem)
    print(company_elem)
    print(location_elem)
    print()
````

Use ```.text.strip()``` to strip all whitespace. We find that there is 1 job which doesn't fit into the structure.

````python
for job_elem in job_elems:
    # Each job_elem is a new BeautifulSoup object.
    # You can use the same methods on it as you did before.
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    if None in (title_elem, company_elem, location_elem):
        continue
    print(title_elem.text.strip())
    print(company_elem.text.strip())
    print(location_elem.text.strip())
    print()
````

````python
for job_elem in job_elems:
    # Each job_elem is a new BeautifulSoup object.
    # You can use the same methods on it as you did before.
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    if None in (title_elem, company_elem, location_elem):
        print(job_elem)
        print()
        continue
    print(title_elem.getText().strip())
    print(company_elem.getText().strip())
    print(location_elem.getText().strip())
    print()
````

We find that it is an ad "searchResultsAds". The ad made it necessary to add an `if None in` statement to skip over it.

## 2.5 Find specific entry
Let's try to find "Senior Solutions Engineer".

````python
senior_jobs = results.find_all('h2', string='Senior Solutions Engineer')
````

````python
print(senior_jobs)
````

This is because of differences in whitespace / capitalisations, since ```string=``` searches for exact matches.

````python
senior_jobs = results.find_all('h2', string=lambda text: 'senior' in text.lower())
````

````python
print(len(senior_jobs))
````

## 2.6 Extract hyperlinks
Use `('a')['href']` notation to access the `href` attribute (hyperlink) and add to our scrap results:

````python
for senior_job in senior_jobs:
    link = senior_job.find('a')['href']
    print(senior_job.text.strip())
    print(f"Apply here: {link}\n")
````
