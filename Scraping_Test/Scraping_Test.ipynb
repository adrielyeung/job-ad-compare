## 2.7 Extract description by accessing another page
We want to extract a detailed description of each job. As a human, we would click on the hyperlink, by doing so, we notice that the URL has an extra https://www.monster.com/jobs/search/?q=Software-Developer&where=hong-kong& *jobid=fd737331-5b02-41ad-9d63-b3719a7bed40* part appended

We discover that the job ID can be extracted from the `section class='card-content'` tag.

````python
options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")

driver = webdriver.Chrome(options=options, executable_path=DRIVER_PATH)
for job_elem in job_elems:
    # Each job_elem is a new BeautifulSoup object.
    # You can use the same methods on it as you did before.
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    if job_elem.find('a') is None:
        continue
    URL_elem = job_elem.find('a')['href']
    
    if None in (title_elem, company_elem, location_elem):
        continue
    try:
        page_content = driver.get(URL_elem)
        content = driver.find_element_by_name('job_description')
        description_elem = content.text
    except KeyError:
        continue
    print("Title: " + title_elem.text.strip())
    print("Company: " + company_elem.text.strip())
    print("Location: " + location_elem.text.strip())
    print("Description")
    print("===========")
    print(description_elem)
    print("Apply here: " + URL_elem)
driver.quit()
````

We find that the `<div id="JobPreview>` does not contain any content, contrary from what we see on the browser. This is due to the use of Javascript in rendering the page. Let's use Selenium to see if we can load those Javascript.

````python
options = Options()
options.headless = True
options.add_argument("--window-size=1920,1200")

driver = webdriver.Chrome(options=options, executable_path=DRIVER_PATH)
description_list = []
for job_elem in job_elems:
    # Each job_elem is a new BeautifulSoup object.
    # You can use the same methods on it as you did before.
    try:
        URL_content = 'https://www.monster.com/jobs/search/?q=Software-Developer&where=hong-kong&jobid=' + job_elem['data-jobid']
        page_content = driver.get(URL_content)
        content = driver.find_element_by_id('JobDescription')
        description_elem = content.text
    except KeyError:
        continue
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    
    print("Title: " + title_elem.text.strip())
    print("Company: " + company_elem.text.strip())
    print("Location: " + location_elem.text.strip())
    print("Description")
    print("===========")
    print(description_elem)
    description_list.append(description_elem)
driver.quit()
````

## Keyword extraction
We now aim to extract the most used keyword considering all the job postings. This will hopefully give us some idea as to the most important skills associated with a certain role.

````python
import re
def pre_process_draft(text):
    # Convert to lowercase
    text = text.lower()
    # Remove special characters (anything other than letters)
    text = re.sub('(\\d|\\W)+',' ', text)
    return text
````

````python
print(pre_process(description_list[-2]))
````

Potential problems are:
1. Skills like C++, C# may be trimmed as C, and Node.js changed to node js.
2. Some words with apostrophes / dots in between, e.g. world's / U.S., are separated into 2 words.

Let's try retaining characters + and #, and only remove full stop if the next char is space.

````python
import re
def pre_process(text):
    # Convert to lowercase
    text = text.lower()
    # Remove special characters that are of no use
    # Mind the apostrophe ’ is not the same as the quote '
    text = re.sub('(\\d|[^+#.’\\w\'])+',' ', text)
    # Remove full stop (recognised by space after it or end of string)
    text = re.sub('\\.\s', ' ', text)
    text = re.sub('\\.$', ' ', text)
    text = re.sub('’','\'', text)
    return text
````

**Stop words** are words which are filtered out before or after processing of natural language data (text), normally the most commonly used words in the language.

We use a stop word list as provided [here](https://github.com/kavgan/nlp-in-practice/tree/master/tf-idf/resources), with minor modifications from experiments.

````python
from sklearn.feature_extraction.text import CountVectorizer
import re

def get_stop_words(stopwords_file_path):
    # load stop words
    with open(stopwords_file_path, 'r', encoding="utf-8") as f:
        stopwords = f.readlines()
        stop_set = set(m.strip() for m in stopwords)
        return frozenset(stop_set)
        
# load stopwords
stopwords = get_stop_words("stopwords.txt")
````

````python
print(stopwords)
````

Now, let's use Scikit-Learn's [CountVectorizer](http://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html) to analyse the job description, which will return a vocabulary list with the counts of each word in the list.

````python
count_vectorizer = CountVectorizer(max_df=0.85, stop_words = stopwords)
````

```max_df``` is a percentage, meaning to filter stop words which have appeared in more than this proportion of documents.

````python
word_count_vector = count_vectorizer.fit_transform(description_list)
````
